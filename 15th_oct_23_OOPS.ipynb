```python
#Constructor :
#10. How can you have multiple constructors in a Python class? Explain with an example.


''' Yes we can have multiple constructor in python using the combinations of default arguments,class methods
    -python does not support method overloading. How ever you can achive the same working effect
     using default arguments.
    -So there are two ways to create the multiple constructor. 
    1. By defining the constructor as default arguments.which allows you to call the constructor by different sets of arguments. 
    2. and by using the decorator "@classmethod",You can create class methods that act as alternative constructors. 
       These methods create and return instances of the class.
'''



class Sample:
    
    def __init__(self,a = 3, b = 6,c = "coder"):
        self.a = a
        self.b = b
        self.c = c
        
    def show(self):
        print(f"a = {self.a}\nb = {self.b}\nc = {self.c}")
        
class MyClass:
    
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    
    @classmethod
    def Constructor_1(cls,a,b):
        return cls(a,b,"Constructor_1")
    
    @classmethod
    def Constructor_2(cls,a):
        return cls(a,"Constructor_2","Hustler")
    
    def show(self):
        print(f"a = {self.a}\nb = {self.b}\nc = {self.c}")

    
print("Using Default Arguments:")
print("OBJECT 1")
obj = Sample(1,4,"hustler")
obj.show()
print("\nOBJECT 2")
obj2 = Sample(1,5)
obj2.show()
print("\nOBJECT 3")
obj3 = Sample("Sagar")
obj3.show()
    

print("\n\nUsing Classmethod decorator:") 
print("OBJECT 1")
obj4 = MyClass(5,7,3)
obj4.show()
print("\nOBJECT 2")
obj5 =MyClass.Constructor_2(6)
obj5.show()
print("\nOBJECT 3")
obj6 = MyClass.Constructor_1("Sudh",2)
obj6.show()
```

```python
#Inheritance
#1. What is inheritance in Python? Explain its significance in object-oriented programming.

'''
    -In OOPS the inheritance is the fundamental concept.
    -That allows the new class to access the functionality and attributes of the old classes.
    -Which enables the reusability of methods and also able to extend the functionality of existing in new class.
    -So this enables the heirarchy of the class creation and access the functionality and datas of all classes by single object.
    -So thats why the Inheritance is the most important part of oops.
'''


class ParentClass:
    
    def __init__(self,a,b):
        self.a = a
        self.b = b
        print("ParentClass Constructor called..")
        
        
    def square(self):
        print("ParentClass Function Called")
        self.a_Sqr = self.a**2
        self.b_Sqr = self.b**2
        return self.a_Sqr + self.b_Sqr
        
        
class ChildClass(ParentClass):
    
       def __init__(self,a,b,c):
            print("ChildClass Constructor called..")
            super().__init__(a,b)
            self.c = c
      
       def productAll(self):
            print("ChildClass Function Called")
            self.sqrProduct = super().square()
            return self.sqrProduct * self.c

#So here we are not calling the data of parent calss.
obj = ChildClass(4,2,8)
print(obj.productAll())
```

```python
#Inheritance
#5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.

'''
    -So in inheritance, sometimes we required to access the functions and the attributes of parent class in the derrived class.
    -So to access the variables and functions of the parent class in child class we have to use the super() inbuilt function.
    -By using the super() function we can access the function and attributes of the parent class.
'''

class Parent:
    
    _class_variable = 1000
    
    def __init__(self,a):
        self.a = a
        
    def add(self,a,b):
        return a+b
    
    
class Child(Parent):
    
    
    def __init__(self,a,b):
        #Accessing the parent class constructor
        super().__init__(a)
        self.b = b
        
    def calc(self):
        #Accessing the parent class function
        self.sum = super().add(self.a,self.b)
        self.product = self.a * self.b
        
    def showdata(self):
        #Accessing the parent class variables.
        print(f"ParentClass variable = {super()._class_variable}")
        print(f"a = {self.a} & b = {self.b}")
        print(f"Sum = {self.sum}")
        print(f"Product = {self.product}")
    
    
obj = Child(5,3)
obj.calc()
obj.showdata()
    

```

```python
#Inheritance
#12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance.
#    Provide an example using the `abc` module.

'''
    -An Abstract base class is a class which contains the abstract method means a method which doesn't have any implementations
     and those methods should be define in the derrived class.
    -But while defining the we doesnt require the super() function to define and thats why its is called as Abstract method.
     So we are not extending the same function in derrived class we have to complete the implementation of that 
     abstract class which is declared in the base class.
    -But in base that method should be decorated by the '@abstractmethod' decorator, which indicates that this method is '@abstractmethod'
     and we have to complete its implementation in the derrived class.
    -The 'abc' module provides 'ABC' class and the '@abstractmethod' decorator..
    -So before defining the abstract method we have to import the 'ABC' class and '@abstractmethod' decorator from 'abc' module.
    -And the class which contains the '@abstractmethod' that class should inherit the class 'ABC'. So after that we can 
     declare the abstract method.
    -And one advantage is that we can define it with different implementation in different classes if the perticular 
     abstract class is inherited by multiple classes.
    -Without using these 'abc' module , class 'ABC' and '@abstractmethod' decorator if we try to declare a method in the class and in implementation
     by write a 'pass', after that try to completely define it in the derrived class, then that method will be overiding the base class method.
     and if we use the super() then it is extending the functionality of base class function and that scenario is not called as a 
     Abstract method scenario.
     
'''

from abc import ABC, abstractmethod

# Defining an abstract base class (ABC)
class Shape(ABC):

    #The are the abstract methods
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius
    
    #Its not overiding and also not require the super() function 
    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius


class Square(Shape):

    def __init__(self, side_length):
        self.side_length = side_length

    def area(self):
        return self.side_length * self.side_length

    def perimeter(self):
        return 4 * self.side_length


circle = Circle(5)
square = Square(4)


print("Circle Area:", circle.area())         
print("Circle Perimeter:", circle.perimeter()) 

print("Square Area:", square.area())          
print("Square Perimeter:", square.perimeter()) 


```

```python
#Encapsulation
#12. Explain the concept of property decorators in Python and how they relate to encapsulation.

'''
    -property decorators are a way to define properties on a class,
    -Which allows you to control access to attributes while providing customization through getter, setter, and deleter methods.
    -They are used to manage and control the access, modification, and deletion of class attributes, 
     providing a level of encapsulation by allowing you to define specific behavior when accessing or modifying an attribute.
    -The property decorators allows you to access a method as an attribute without using parentheses.
    
    -@property: This decorator is used to define a property and is placed before a method in a class. 
                Which is used to get the data.
    -@<attribute_name>.setter: This decorator is used to define the setter method for the corresponding property.
                It allows you to modify the property value.           
    -@<attribute_name>.deleter: This decorator is used to define the deleter method for the corresponding property.
                It allows you to perform actions when the property is deleted.  
                
    -You can perform this operation with private  and protected variables and attributes.           
'''

class Sample:
    
    def __init__(self,name):
        self.__name = name
        
    
    @property
    def name(self):
        return self.__name
    
    
    @name.setter
    def name(self,new_name):
        try:
            if isinstance(new_name,str) and len(new_name)>0:
                self.__name = new_name
                return f"Value updated.."
                
            else:
                raise ValueError("The value must be non empty string..")
        except Exception as e:
            return e

        
    @name.deleter
    def name(self):
        print("Deleting the attribute..After this if you try to fetch this attribute it will raise the AttributeError..")
        del self.__name
        
#Creating the instance by passing the valid name..
obj = Sample("Ashish")
#Fetching the attribute using the getter method.
print(obj.name)

#Updating the name using setter method
obj.name = "Prashant"
print(obj.name)

#Deleting attribute using the deleter method.
del obj.name
#After deleting the attribute we are trying to fetch. but it is deleted then it will raise the attribute error
print(obj.name)
```

```python
#Encapsulation
#17. Create a Python class for a library system that encapsulates book information, including titles, authors, and availability status.


class Library:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

    def __str__(self):
        availability = "Available" if self.available else "Not Available"
        return f"Title: {self.title}\nAuthor: {self.author}\nAvailability: {availability}"

    def borrow_book(self):
        if self.available:
            self.available = False
            print(f"The book '{self.title}' has been borrowed.")
        else:
            print(f"Sorry, the book '{self.title}' is not available.")

    def return_book(self):
        if not self.available:
            self.available = True
            print(f"Thank you for returning the book '{self.title}'.")
        else:
            print("This book is already available in the library.")



if __name__ == "__main__":
   
    book1 = Library("Python Crash Course", "Eric Matthes")
    book2 = Library("The Alchemist", "Paulo Coelho", available=False)

    
    print("Book 1:")
    print(book1)
    print("\nBook 2:")
    print(book2)

    book1.borrow_book()
    book2.borrow_book()
    book2.return_book()

    
```

```python
#Polymorphism:-
#3. Create a Python class hierarchy for shapes (e.g., circle, square, triangle) and demonstrate polymorphism through
#   a common method, such as `calculate_area()`.

class Shapes:
    
    def calculate_area(self,radius = None,side_length = None, base = None, height = None):
        if radius != None:
            if radius > 0:
                return 3.14*radius**2
        elif side_length != None:
            if side_length>0:
                return side_length**2
        if base is not None and height is not None and base > 0 and height > 0:
            if base > 0 and height > 0:
                return 0.5 * base * height
        else:
            return f"Invalid passed value"
        
class Circle(Shapes):
    
    def __init__(self,radius):
        self.radius = radius
        
    def calculate_area(self):
        return f"Area of circle of radius {self.radius} : {super().calculate_area(self.radius)}"
    
class Square(Shapes):
    
    def __init__(self,side_length):
        self.side_length = side_length
        
    def calculate_area(self):
        return f"Area of Square with side length {self.side_length} : {super().calculate_area(self.side_length)}"
    
class Triangle(Shapes):
    
    def __init__(self,base,height):
        self.base = base
        self.height = height
        
    def calculate_area(self):
        return f"Area of Triangle with base {self.base} and height {self.height} : {super().calculate_area(self.base,self.height)}" 
    
    
circle = Circle(6)
print(circle.calculate_area())

square = Square(5)
print(square.calculate_area())

triangle = Triangle(7,8)
print(triangle.calculate_area())
```

```python
#Abstraction:
#15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.


'''
    -Abstract DataType:
    -So the abstarct datatype is a part of the theorotical concept if the abstract classes.
    -Whenever we define the abstract class it means that we are defining the data types and irs behavior of the objects which we will be created.
    -Internally the the abstract class defines the behaviour of the data rather than its implementations.
    -Abstract data types is essential in python to achive the abstraction.
    
    -Abstarction is used for hiding the complex implementations by defining the abstract methods and expose only those which is necessary 
     features and functionality.
    -So the Abstract data types facilitate to programmer to access and work with data without knowing the internal implementation 
     of the abstract class.
    -ADTs define a set of operations or behaviors that can be performed on a data type. 
    -ADTs allow multiple implementations for the same abstract data type.
    -ADTs encapsulate data and operations, providing a clear interface for users.
    -Users interact with the ADT through its defined operations, shielding them from the internal details.
    

'''


class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)  

    def peek(self):
        if not self.is_empty():
            return self.items[0]  
        
        
obj = Queue()

print(obj.is_empty())
obj.enqueue(6)
obj.enqueue(9)
obj.enqueue("coder")
obj.enqueue(16)
obj.enqueue(60.6)
print(obj.is_empty())
print(obj.items)
print(obj.dequeue())
print(obj.peek())
print(obj.items)
```

```python
#Abstraction:
#20. Describe the concept of method abstraction in Python and how it relates to polymorphism.

'''
    -Abstract methods are defined using the @abstractmethod decorator from the abc module in Python. 
     These methods do not have an implementation within the abstract class; instead, they are meant to be overridden by subclasses. 
     An abstract class itself cannot be instantiated; it serves as a blueprint for creating concrete classes that implement its abstract methods.
    -Polymorphism, on the other hand, refers to the ability of different classes to be treated as objects of a common superclass. In Python, 
     polymorphism allows objects of different classes to be used interchangeably if they inherit from the same abstract class or share a common interface. 
    -method abstraction allows us to define methods without implementation details in abstract classes, and polymorphism enables the interchangeability 
     and uniform treatment of objects that adhere to a common interface or superclass, even if they have different implementations for the methods.

'''

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius * self.radius


rectangle = Rectangle(5, 10)
print("Area of Rectangle:", rectangle.area())  # Output: 50

circle = Circle(7)
print("Area of Circle:", circle.area())  # Output: 153.86
```

```python
#Composition
#2. Describe the difference between composition and inheritance in object-oriented programming

'''
    -composition and inheritance are two fundamental concepts used to establish relationships between classes,
     but they differ in their approach to building complex structures and code reuse.
     
    -Composition:
    -Composition establishes a "has-a" relationship between classes. It allows a class to contain objects of other classes as components or parts.
    -The containing class holds references to instances of other classes as attributes, utilizing their functionalities.
    -Composition promotes encapsulation by allowing objects to manage their own behavior and data independently. Each object maintains its state and
     functionality, enhancing modularity and code reuse.
    -Composition provides flexibility in creating complex structures. It enables classes to be composed and recomposed, allowing for more dynamic
     and flexible relationships between objects.
    
    
    -Inheritance:
    - Inheritance establishes an "is-a" relationship between classes. It allows a subclass to inherit properties and behavior from a superclass.
    -It involves creating new classes (subclasses) based on existing classes (superclasses), inheriting their attributes and methods. Subclasses 
     can extend or override the functionality of their parent classes.
    -Inheritance promotes code reuse by allowing subclasses to inherit and use the attributes and methods of their parent classes. 
     It enables the extension of functionality without rewriting code.
    -Inheritance forms a hierarchical model, defining an "is-a" relationship where subclasses are more specialized versions of their parent classes.
'''

#Composition Example:
class Engine:
    def start(self):
        print("Engine started.")

    def stop(self):
        print("Engine stopped.")

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition: Car has an Engine

    def start(self):
        print("Car started.")
        self.engine.start()

    def stop(self):
        print("Car stopped.")
        self.engine.stop()
        
        
class Bike(Engine):       #Inheritance: Inheriting the Engine class
    
    def start(self):
        print("Bike started.")
        super().start()

    def stop(self):
        print("Bike stopped.")
        super().stop()
        
#Composition Eg        
car = Car()
car.start()
car.stop()
print()

#Inheritance Eg
bike = Bike()
bike.start()
bike.stop()
```

```python
#Composition
#5. How can you implement composition in Python classes? Provide examples of using composition to create complex objects.

'''
    -In Python, composition can be implemented by creating classes that contain instances of other classes as attributes. 
    -Following example illustrate how composition allows the creation of complex objects by combining simpler objects or components,
     providing flexibility, code reuse, and modularity in designing object-oriented systems.
'''

class CPU:
    def process(self):
        print("CPU processing...")

class RAM:
    def load(self):
        print("RAM loading data...")

class GraphicsCard:
    def render(self):
        print("Graphics card rendering...")

class Computer:
    def __init__(self):
        self.cpu = CPU()  
        self.ram = RAM()  # Composition: Computer has RAM
        self.graphics_card = GraphicsCard()  # Composition: Computer has a GraphicsCard

    def run_program(self):
        print("Running a program...")
        self.cpu.process()
        self.ram.load()
        self.graphics_card.render()

# Creating a Computer object
my_computer = Computer()

# Using Computer object
my_computer.run_program()
```

```python
#Composition
#12. Create a Python class for a university course, using composition to represent students, instructors, and course materials.

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def __str__(self):
        return f"Student ID: {self.student_id}, Name: {self.name}"

class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def __str__(self):
        return f"Instructor ID: {self.instructor_id}, Name: {self.name}"

class CourseMaterial:
    def __init__(self, material_name):
        self.material_name = material_name

    def __str__(self):
        return f"Course Material: {self.material_name}"

class UniversityCourse:
    def __init__(self, course_name, instructor, students, course_materials):
        self.course_name = course_name
        self.instructor = instructor
        self.students = students
        self.course_materials = course_materials

    def get_course_details(self):
        details = f"Course: {self.course_name}\n"
        details += f"Instructor: {self.instructor}\n"
        details += "Students:\n"
        for student in self.students:
            details += f"- {student}\n"
        details += "Course Materials:\n"
        for material in self.course_materials:
            details += f"- {material}\n"
        return details


student1 = Student(1, "Ashish")
student2 = Student(2, "Nishant")

instructor = Instructor(101, "Dr. Fandat")

material1 = CourseMaterial("Textbook: Introduction to Python")
material2 = CourseMaterial("Lecture Slides: Python Basics")

# Creating a UniversityCourse instance
course = UniversityCourse("Python Programming", instructor, [student1, student2], [material1, material2])


course_details = course.get_course_details()
print(course_details)

```

```python
#Composition
#13. Discuss the challenges and drawbacks of composition, such as increased complexity and potential for tight coupling between objects.

'''
    -A composition is a powerful design principle that promotes code reusability, modularity, and flexibility.
     it also introduces certain challenges and drawbacks.
     
    -Increased Complexity: As more objects are composed together to build complex structures, the overall system's 
     complexity can increase. Managing multiple interconnected objects and their interactions might become more challenging, 
     especially in larger systems, leading to code that is harder to understand, debug, and maintain.
     
    -Tight Coupling: Composition can lead to tight coupling between objects, where the functionality of one object depends heavily
     on the implementation details of another. If the composed objects are highly interdependent, changes in one object might require 
     modifications in multiple other objects, making the system less flexible and more prone to cascading changes. 
     
    -Dependency Management: When objects are composed, managing dependencies between them becomes crucial. Changes in the behavior or
     structure of one object might affect others. This makes it essential to carefully design interfaces and contracts between objects to 
     minimize the impact of changes and prevent unintended consequences.
    
    -Inheritance vs. Composition: In some cases, choosing between inheritance and composition can be challenging. While composition promotes 
     flexibility by favoring object composition over inheritance, deciding the right level of abstraction and the best way to structure the 
     relationships between objects can be a complex decision.
     
    -Performance Overhead: Depending on the implementation, composition might introduce performance overhead, especially if there are numerous 
     object interactions or if objects are too granular. Managing these interactions efficiently without causing performance 
     bottlenecks can be a challenge.
    
    -Complex Initialization: As more objects are composed together, the initialization and setup of these objects can become more intricate. 
     This complexity in object creation might lead to issues with object instantiation and initialization order, especially in scenarios with deep
     object hierarchies.
     
     
    -Use design patterns like the Factory pattern, Dependency Injection, or the Facade pattern to manage object creation, dependencies, and
     interactions effectively.
    -Implement clear interfaces between composed objects to reduce tight coupling and dependency on specific implementations.
    -Apply principles such as the Single Responsibility Principle (SRP) and Separation of Concerns (SoC) to ensure that each object has a clear, 
     well-defined purpose and doesn't handle too many responsibilities.
    -Regularly refactor and review the codebase to simplify complex compositions and reduce unnecessary dependencies.
'''
```

```python
#Composition
#16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.

class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def __str__(self):
        return f"Weapon: {self.name}, Damage: {self.damage}"

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def __str__(self):
        return f"Armor: {self.name}, Defense: {self.defense}"

class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def __str__(self):
        inventory_info = "Inventory:\n"
        for item in self.items:
            inventory_info += f"- {item}\n"
        return inventory_info

class GameCharacter:
    def __init__(self, name):
        self.name = name
        self.weapon = None  
        self.armor = None  
        self.inventory = Inventory()

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def add_item_to_inventory(self, item):
        self.inventory.add_item(item)

    def __str__(self):
        character_info = f"Character: {self.name}\n"
        if self.weapon:
            character_info += f"Weapon Equipped: {self.weapon}\n"
        if self.armor:
            character_info += f"Armor Equipped: {self.armor}\n"
        character_info += f"{self.inventory}"
        return character_info

# Creating weapons, armor, and items for the character
sword = Weapon("Sword", 20)
shield = Armor("Shield", 15)
potion = "Health Potion"


hero = GameCharacter("Hero")
hero.equip_weapon(sword)
hero.equip_armor(shield)
hero.add_item_to_inventory(potion)

print(hero)
```

```python
#Composition
#18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.

class Furniture:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Furniture: {self.name}"

class Appliance:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Appliance: {self.name}"

class Room:
    def __init__(self, name):
        self.name = name
        self.furniture = []
        self.appliances = []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def __str__(self):
        room_info = f"Room: {self.name}\n"
        room_info += "Furniture:\n"
        for furniture_item in self.furniture:
            room_info += f"- {furniture_item}\n"
        room_info += "Appliances:\n"
        for appliance_item in self.appliances:
            room_info += f"- {appliance_item}\n"
        return room_info

class House:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)

    def __str__(self):
        house_info = "House Details:\n"
        for room in self.rooms:
            house_info += f"{room}\n"
        return house_info

# Creating furniture and appliances for rooms
bed = Furniture("Bed")
sofa = Furniture("Sofa")
tv = Appliance("Television")
fridge = Appliance("Fridge")

# Creating rooms and adding furniture and appliances
bedroom = Room("Bedroom")
bedroom.add_furniture(bed)
living_room = Room("Living Room")
living_room.add_furniture(sofa)
living_room.add_appliance(tv)
kitchen = Room("Kitchen")
kitchen.add_appliance(fridge)


my_house = House()
my_house.add_room(bedroom)
my_house.add_room(living_room)
my_house.add_room(kitchen)

print(my_house)
```

```python
#Composition
#19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?


'''
    -Achieving flexibility in composed objects to allow dynamic replacement or modification at runtime can be accomplished using
     various design patterns and programming techniques. Here are a few strategies:
    -Dependency Injection (DI):Dependency Injection allows objects to be passed as dependencies rather than creating them within 
     the class. This technique enables dynamic replacement or modification of dependencies by injecting different objects at runtime.
     
    -Strategy Pattern:
    -The Strategy pattern defines a family of algorithms, encapsulates each one, and makes them interchangeable. It allows the behavior
     of an object to be selected at runtime.
     
    -Factory Pattern:
    -The Factory pattern encapsulates object creation. It allows the creation of objects without specifying the exact class of the object that 
     will be created, enabling dynamic object creation at runtime.
     
    -techniques like Dependency Injection, Strategy Pattern, or Factory Pattern, you can create composed objects that allow for dynamic 
     replacement or modification of their components at runtime, enhancing flexibility and adaptability in your codebase.
'''


class SortStrategy:
    def sort(self, data):
        pass

class QuickSort(SortStrategy):
    def sort(self, data):

        print("Sorting using Quick Sort")

class MergeSort(SortStrategy):
    def sort(self, data):

        print("Sorting using Merge Sort")

class Sorter:
    def __init__(self, strategy):
        self.strategy = strategy

    def set_strategy(self, new_strategy):
        self.strategy = new_strategy

    def sort_data(self, data):
        return self.strategy.sort(data)

# Creating a sorter with a default strategy
default_strategy = QuickSort()
sorter = Sorter(default_strategy)

# Sorting data using the default strategy
data = [5, 2, 9, 1, 5]
sorter.sort_data(data)

# Later, changing the sorting strategy dynamically
new_strategy = MergeSort()
sorter.set_strategy(new_strategy)

# Sorting data using the new strategy
sorter.sort_data(data)
```

```python
#Composition
#20. Create a Python class for a social media application, using composition to represent users, posts, and comments.
        
    
class Comment:
    def __init__(self, text, author):
        self.text = text
        self.author = author

    def __str__(self):
        return f"Comment: {self.text} by {self.author}"

class Post:
    def __init__(self, content, author):
        self.content = content
        self.author = author
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def __str__(self):
        post_info = f"Post: {self.content} by {self.author}\nComments:\n"
        for comment in self.comments:
            post_info += f"- {comment}\n"
        return post_info

class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def create_post(self, content):
        post = Post(content, self.username)
        self.posts.append(post)
        return post

    def add_comment_to_post(self, post, text):
        comment = Comment(text, self.username)
        post.add_comment(comment)

    def __str__(self):
        user_info = f"User: {self.username}\nPosts:\n"
        for post in self.posts:
            user_info += f"{post}\n"
        return user_info

class SocialMediaApp:
    def __init__(self):
        self.users = []

    def create_user(self, username):
        user = User(username)
        self.users.append(user)
        return user


social_media = SocialMediaApp()


user1 = social_media.create_user("Alice")
user2 = social_media.create_user("Bob")

post1 = user1.create_post("Hello, everyone!")
user1.add_comment_to_post(post1, "Nice post, Alice!")

post2 = user2.create_post("Python is awesome!")
user2.add_comment_to_post(post2, "I totally agree!")


print(user1)
print(user2)
    
```
